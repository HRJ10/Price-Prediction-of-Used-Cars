```python
import numpy as np 
import pandas as pd     

import warnings         
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("dataset/clean-train.csv")
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

cat_cols = ['Fuel', 'Transmission', 'Owner']

encoder = OneHotEncoder(sparse_output=False)
encoded_cols = encoder.fit_transform(df[cat_cols])

encoded_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(cat_cols))

dfe = pd.concat([df.drop(cat_cols, axis=1), encoded_df], axis=1)

dfe = dfe.dropna()

dfe.head()
```

```python
dfe.to_csv('dataset/encoded-train.csv', index=False)
```
